```python
# Por cada elemento de la lista comprueba si solo contiene caracteres alfabéticos, minusculas o digitos.


text_entries = ['Hola', 'hola', '1234', "hola1234"]  # Lista de entradas de texto

for text in text_entries:

    print(f"Texto '{text}':")

    # Comprueba si el texto solo contiene caracteres alfabéticos
    # escribe aquí tu función de predicado
    condicion_1 = text.isalpha()
    print("Solo contiene alfabeto", condicion_1)

    # Comprueba si el texto contiene solo letras en minúscula
    # escribe aquí tu función de predicado
    condicion_2 = text.islower()
    print("Solo contiene minúsculas", condicion_2 )

    # Comprueba si el texto contiene solo dígitos
    # escribe aquí tu función de predicado
    condicion_3 = text.isdigit()
    print("Solo contiene dígitos", condicion_3)
```

```python
# Misma cadena, pero ahora con condiciones diferentes

text_entries = ['Hola', 'hola', '1234', "hola1234"]  # Lista de entradas de texto

for text in text_entries:

    print(f"Texto '{text}':")

    # Comprueba si el texto es todo alfabeto o son todos dígitos
    # escribe aquí tus funciones de predicado con la expresión lógica
    condicion_1 = text.isalpha() or text.isdigit()
    print("Es alfabético o tiene dígitos", condicion_1)

    # Comprueba si el texto no es todo alfabeto o no son todos dígitos
    # escribe aquí tus funciones de predicado con la expresión lógica
    condicion_2 = not (text.isalpha() or text.isdigit())
    print("No es alfabético ni son dígitos", condicion_2)
```

# Modificación de tablas

```python
# Reemplazar la película a cambiar por la nueva mediate un ciclo for

movies = [
    ["The Shawshank Redemption", 1994, "Frank Darabont"],
    ["The Godfather", 1972, "Francis Ford Coppola"],
    ["The Dark Knight", 2008, "Christopher Nolan"],
    ["12 Angry Men", 1957, "Sidney Lumet"],
    ["Schindler's List", 1993, "Steven Spielberg"],
    ["The Lord of the Rings: The Return of the King", 2003, "Peter Jackson"]
]

movie_to_change = "The Lord of the Rings: The Return of the King"
new_movie = "The Lord of the Rings: The Fellowship of the Ring"
new_year = 2001

for movie in movies:
# establece una condición aquí
    if movie[0] == movie_to_change:
      #sustituye el nombre de la película si la condición devuelve True
      movie[0] = new_movie
      # sustituye el año de estreno si la condición devuelve True
      movie[1] = new_year

# no modifiques el código de abajo, ya que imprime el resultado
for movie in movies:
    print(movie)
```
